### Removendo NaNs de Arrays

```python
x4 = np.array([1, 2, np.nan, 3, np.nan, 5, np.nan, 2, 1, 5, 8, 9, np.nan, 11, 3, 2, 7, np.nan])
```

```python
lista = []
for i in x4:
  if np.isnan(i):
    print('---> NaN')
  else:
    lista.append(i)
    print(i)
```

```python
x5 = np.array(lista)
```

```python
x5
```

```python
x6 = x4[~np.isnan(x4)]
```

```python
x5
```

### Operações em vetores com NaN

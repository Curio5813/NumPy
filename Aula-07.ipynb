## Técnicas avançadas de Manipulação de Arrays 2D

```python
np.random.seed(2)
```

```python
x = np.random.rand(3, 3); x
```

```python
y = x.copy() + 1; y
```

```python
z = y[:,2].copy(); z
```

```python
# Empilha verticalmente

np.vstack([z,z,z,z])
```

```python
np.array([z,z,z,z])
```

```python
# Função np.reshape e np.hstack
```

```python
z.T
```

```python
z1 = np.reshape(z, (3,1) ); z1
```

```python
# Empilha horizontalmente
```

```python
np.hstack([z1,z1,z1,z1])
```

```python
# Avaliações booleanas básicas
```

```python
x = np.array([1.0, 3. , 2.6, 4.3, 7.7, 5.6])
y = np.array([1. , 2.6 , 3., 4. , 7. , 5.6])
```

```python
x
```

```python
y
```

```python
#x.shape[0]
len(y)
```

```python
cmp = x == y
```

```python
cmp
```

```python
cmp.sum()
```

```python
if(cmp.sum() == len(x)):
  print('Os vetores são iguais')
else:
  print('Os vetores são diferentes')
```

```python
Z1 = np.array([1.0, 3. , 2.6, 4.3, 7.7, 5.6])
Z2 = np.array([1.0, 3. , 2.6, 4.3, 7.7, 5.6])
```

```python
cmp2 = Z1 == Z2
```

```python
cmp2
```

```python
if(cmp2.sum() == len(Z1)):
  print('Os vetores são iguais')
else:
  print('Os vetores são diferentes')
```
